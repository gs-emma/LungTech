```python
import librosa
import tensorflow as tf
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import numpy.ma as ma 
import pandas as pd
import pickle

import warnings
warnings.filterwarnings ('ignore')
```

```python
# reading the csv file
data = pd.read_csv("/Users/home/Documents/Respiratory Illness AI/healthy_others_metadata.csv")
data.head(5)
```

```python
# converting all data to spectrograms 128,31

dataset1 = []

for row in data.itertuples():
    y, sr = librosa.load(f"/Users/ashley/segmented_raw_files/copd_healthy/{row.fileName}.wav", sr = 48000)
    
    audio = librosa.feature.melspectrogram(y=y, sr=sr)
    
    if audio.shape == (128, 31):
        dataset1.append( (audio, row.id) )
    print (len(dataset1))
print ("complete")
```

```python
# converting all data to spectrograms 224, 224

dataset2 = []

for row in data.itertuples():
    y, sr = librosa.load(f"/Users/ashley/segmented_raw_files/copd_healthy/{row.fileName}.wav", sr = 347530)
    
    audio = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=224)
    
    if audio.shape == (224, 224):
        dataset2.append( (audio, row.id) )
    print (len(dataset2))
print ("complete")
```

```python
# converting all data to spectrograms 224, 224, 3
dataset3 = []

for row in data.itertuples():
    y, sr = librosa.load(f"/Users/home/Documents/Respiratory Illness AI/recordings/official_training/healthy_others/{row.fileName}.wav", sr = 347530)
    
    audio = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=224)
    audio2 = np.stack((audio,)*3, axis=-1)
    
    if audio2.shape == (224, 224, 3):
        dataset3.append( (audio2, row.id) )
    print (len(dataset3))
print ("complete")
```

```python
# converting all data to spectrograms 229, 229

dataset4 = []

for row in data.itertuples():
    y, sr = librosa.load(f"/Users/ashley/segmented_raw_files/copd_healthy/{row.fileName}.wav", sr = 355000)
    
    audio = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=229)
    
    if audio.shape == (229, 229):
        dataset4.append( (audio, row.id) )
    print (len(dataset4))
print ("complete")
```

```python
# converting all data to spectrograms 229, 229, 3

dataset5 = []

for row in data.itertuples():
    y, sr = librosa.load(f"/Users/ashley/segmented_raw_files/copd_healthy/{row.fileName}.wav", sr = 355000)
    
    audio = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=229)
    audio2 = np.stack((audio,)*3, axis=-1)
    
    if audio2.shape == (229, 229, 3):
        dataset5.append( (audio2, row.id) )
    print (len(dataset5))
print ("complete")
```
